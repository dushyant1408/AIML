```python
# Recreate your own Python NB for “Classes and Objects” from the pdf provided namely
# “Classes&Objects.pdf”

class BankAccount:
    __accountValue = 0 
    
    def OpenAccount(self, name):
        print('Welcome',name)
        __accountValue = 0 

    def __init__(self):
        print("Please enter some Amount")
        
    def AddAmount(self, saveValue):
        self.__accountValue = self.__accountValue + saveValue 
    
    def Withdraw(self, theamount):
        self.__accountValue = self.__accountValue - theamount 
    
    def ShowAmount(self): 
        print('Your Saved Amount is', self.__accountValue)
    
    def __del__(self):
        print('Acocunt has been deleted. Reimbursing amount', self.__accountValue)
    
customer = BankAccount(); 
customer.OpenAccount("My Name")

customer.ShowAmount()
customer.AddAmount(5000)
customer.ShowAmount()
customer.Withdraw(2000)
customer.ShowAmount()

del customer
#customer.
```
